```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
df['Fuel_Type']=le.fit_transform(df['Fuel_Type'])
df['Seller_Type']=le.fit_transform(df['Seller_Type'])
df['Transmission']=le.fit_transform(df['Transmission'])
```

## model training

```python
from sklearn.linear_model import LinearRegression
import xgboost as xgb
from sklearn.model_selection import GridSearchCV
from xgboost import XGBRegressor
from sklearn.metrics import mean_absolute_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
x.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=42,test_size=0.2)
```

## LinearRegression

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
predict=model.predict(x_test)
```

```python
predict
```

```python
predict.shape
```

```python
mae = mean_absolute_error(y_test,predict)
mae
```

## XGB

```python
from sklearn.pipeline import Pipeline
```

```python
pipeline=Pipeline([
    ('clf',xgb.XGBRegressor())
])
param_grid = {
    'clf__learning_rate': [0.01, 0.1, 0.3],  # Step size shrinkage
    'clf__n_estimators': [100, 300, 500],   # Number of boosting rounds
    'clf__max_depth': [3, 5, 7],             # Maximum depth of trees     
    'clf__gamma': [0, 0.1, 0.3],             # Minimum loss reduction
    'clf__subsample': [0.8, 1.0]            # Fraction of samples used
          
}
```

```python
grid_search = GridSearchCV(estimator=pipeline, param_grid=param_grid, 
                           cv=3, scoring='neg_mean_absolute_error', verbose=1, n_jobs=-1)
```

```python
grid_search.fit(x_train,y_train)
```

```python
best_params=grid_search.best_params_
```

```python
best_learning_rate = best_params['clf__learning_rate']
best_n_estimators = best_params['clf__n_estimators']
best_max_depth = best_params['clf__max_depth']
best_gamma = best_params['clf__gamma']
best_subsample = best_params['clf__subsample']
```

```python
model2=xgb.XGBRegressor(learning_rate=best_learning_rate,
    n_estimators=best_n_estimators,
    max_depth=best_max_depth,
    gamma=best_gamma,
    subsample=best_subsample)
```

```python
model2.fit(x_train,y_train)
```

```python
prediction=model2.predict(x_test)
mae=mean_absolute_error(y_test,prediction)
r2score=model2.score(x_test,y_test)
print(f"Test MAE: {mae}")
print("r2_train",model2.score(x_train,y_train))
print(f"R-squared: {r2score}")
```

## Random foerest

```python
from sklearn.ensemble import RandomForestRegressor
regressor=RandomForestRegressor()
param_grid3 = {
    'n_estimators': [100, 200, 300],           # Number of trees in the forest
    'max_depth': [10, 20, 30, None],           # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],           # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4]            # Minimum number of samples required to be at a leaf node
                     
}
```

```python
grid_search3 = GridSearchCV(estimator=regressor, param_grid=param_grid3, cv=5, scoring='neg_mean_squared_error', verbose=1, n_jobs=-1)
grid_search3.fit(x_train,y_train)
```

```python
best_params_=grid_search3.best_params_
```

```python
best_params_
```

```python
best_n_est = best_params_['n_estimators']
best_max_dep = best_params_['max_depth']
best_min_split = best_params_['min_samples_split']
best_leaf = best_params_['min_samples_leaf']
```

```python
regressor=RandomForestRegressor(n_estimators=best_n_est,max_depth=best_max_dep,min_samples_split=best_min_split,min_samples_leaf=best_leaf)
regressor.fit(x_train,y_train)
```

```python
prediction=regressor.predict(x_test)
mae=mean_absolute_error(y_test,prediction)
r2score=regressor.score(x_test,y_test)
print(f"Test MAE: {mae}")
print("r2_train",regressor.score(x_train,y_train))
print(f"R-squared: {r2score}")
```
